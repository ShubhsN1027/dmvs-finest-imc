```python
import pandas as pd
import matplotlib.pyplot as plt

rcParams['figure.figsize'] = 15, 10
```

```python
# read the file called "prices.csv"
prices = pd.read_csv("prices.csv", sep=';')
```

```python
mayberries = prices[prices['product'] == 'BERRIES']
#plot timestamp vs mid_price
plt.plot(mayberries['timestamp'], mayberries['mid_price'])
plt.show()
```

```python
diving_gear = prices[prices['product'] == 'DIVING_GEAR']
dolphin_sightings = prices[prices['product'] == 'DOLPHIN_SIGHTINGS']
```

```python
#add a delta column to dolphin sightings
dolphin_sightings['delta'] = dolphin_sightings['mid_price'].diff()
dolphin_sightings['absdelta'] = dolphin_sightings['delta'].abs()
deltaDolphin = dolphin_sightings.copy()
#sort by absdelta
deltaDolphin.sort_values(by=['absdelta'], inplace=True, ascending=False)
```

```python
#plot diving gear profit_and_loss
plt.plot(diving_gear['timestamp'], diving_gear['profit_and_loss'])
```

```python
plt.plot(dolphin_sightings['timestamp'], dolphin_sightings['mid_price'])
```

```python
plt.plot(diving_gear['timestamp'], diving_gear['mid_price'])
```

```python
# plot dolphin sightings and diving gear vertically stacked on two plots from timestamp 64500 to 66000
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
sub_dolphin = dolphin_sightings[(dolphin_sightings['timestamp'] >= 64500) & (dolphin_sightings['timestamp'] <= 66000)]
sub_diving = diving_gear[(diving_gear['timestamp'] >= 64500) & (diving_gear['timestamp'] <= 66000)]
ax1.plot(sub_dolphin['timestamp'], sub_dolphin['mid_price'])
ax2.plot(sub_diving['timestamp'], sub_diving['mid_price'])
```
